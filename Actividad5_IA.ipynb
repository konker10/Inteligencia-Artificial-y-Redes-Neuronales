### Normalización de Datos Numéricos con Min-Max Scaling

Primero, aplicaremos la normalización Min-Max a las columnas numéricas de nuestro DataFrame `df_cleaned`.

```python
from sklearn.preprocessing import MinMaxScaler

# Identificar las columnas numéricas para la normalización (excluyendo 'Tiempo (s)' si se considera un índice)
numerical_cols = ['Distancia (cm)', 'Velocidad (cm/s)', 'Temperatura (°C)']

# Crear una copia del DataFrame limpio para la normalización
df_normalized = df_cleaned.copy()

# Inicializar el MinMaxScaler
scaler = MinMaxScaler()

# Aplicar la normalización Min-Max a las columnas seleccionadas
df_normalized[numerical_cols] = scaler.fit_transform(df_normalized[numerical_cols])

print("DataFrame después de la normalización Min-Max:")
display(df_normalized.head())
```

### Limpieza de Datos

Primero, mostraremos la cantidad de valores nulos y la forma del DataFrame original antes de empezar la limpieza. Usaremos el DataFrame original `df` para esta demostración.

```python
print("Valores nulos por columna en el DataFrame original:\n", df.isnull().sum())
print(f"Forma del DataFrame original: {df.shape}")
```

Ahora, eliminaremos las filas con cualquier valor nulo del DataFrame original `df`.

```python
df_cleaned_new = df.dropna()
print(f"Forma del DataFrame después de eliminar nulos: {df_cleaned_new.shape}")
```

A continuación, filtraremos los valores atípicos, específicamente las velocidades negativas, ya que la velocidad en este contexto no debería ser negativa. Trabajaremos sobre `df_cleaned_new`.

```python
df_cleaned_new = df_cleaned_new[df_cleaned_new['Velocidad (cm/s)'] >= 0]
print(f"Forma del DataFrame después de filtrar velocidades negativas: {df_cleaned_new.shape}")

display(df_cleaned_new.head())
```

```python
df_cleaned = df.dropna()
print(f"Forma del DataFrame después de eliminar nulos: {df_cleaned.shape}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los gráficos
sns.set_style("whitegrid")

# Crear una figura con dos subplots para comparar antes y después
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histograma de 'Distancia (cm)' antes de la normalización
sns.histplot(df_cleaned['Distancia (cm)'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distancia (cm) - Antes de la Normalización')
axes[0].set_xlabel('Distancia (cm)')
axes[0].set_ylabel('Frecuencia')

# Histograma de 'Distancia (cm)' después de la normalización
sns.histplot(df_normalized['Distancia (cm)'], kde=True, ax=axes[1], color='lightcoral')
axes[1].set_title('Distancia (cm) - Después de la Normalización (Min-Max)')
axes[1].set_xlabel('Distancia Normalizada (0-1)')
axes[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
